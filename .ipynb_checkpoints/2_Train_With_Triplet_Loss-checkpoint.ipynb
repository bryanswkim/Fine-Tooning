```python
import torch
import random
import pickle
import dnnlib
import legacy
import PIL.Image
from PIL import Image
import numpy as np
import torch.nn as nn
from facenet_pytorch import MTCNN, InceptionResnetV1
```

```python
device = torch.device('cuda')
ours_network = "./models/ours_Webtoon.pkl"
with open(ours_network,'rb') as f:
    G = pickle.load(f).to(device)
    
ours_toon_network = "./models/ours_Webtoon_negative.pkl"
with open(ours_toon_network,'rb') as f:
    G_toon = pickle.load(f).to(device)
    
network_pkl = 'https://nvlabs-fi-cdn.nvidia.com/stylegan2-ada-pytorch/pretrained/transfer-learning-source-nets/ffhq-res256-mirror-paper256-noaug.pkl'
with dnnlib.util.open_url(network_pkl) as f:
    human_G = legacy.load_network_pkl(f)['G_ema'].to(device)
    
face_encoder = InceptionResnetV1(pretrained='vggface2').eval().to(device)
```

```python
for name,param in G.named_parameters():
    param.requires_grad =True

for name,param in G_toon.named_parameters():
    param.requires_grad =False    
    
for name,param in face_encoder.named_parameters():
    param.requires_grad =False
```

```python
# Hyperparameters
class_idx = None
truncation_psi = 0.7

label = torch.zeros([1, G.c_dim], device=device)
if G.c_dim != 0:
    if class_idx is None:
        ctx.fail('Must specify class label with --class when using a conditional network')
    label[:, class_idx] = 1
else:
    if class_idx is not None:
        print ('warn: --class=lbl ignored when running on an unconditional network')
```

```python
# function for visualizing image

def vis_image (img):
    temp_img = (img.permute(0, 2, 3, 1) * 127.5 + 128).clamp(0, 255).to(torch.uint8)
    vis_img = PIL.Image.fromarray(temp_img[0].cpu().numpy(), 'RGB')
    vis_img.show()
```

```python
# Loss and optimizer

triplet_loss = nn.TripletMarginWithDistanceLoss(distance_function=nn.CosineSimilarity())
optimizer = torch.optim.Adam(G.parameters(), lr = 1e-5)
epochs = 10

print("Ready!")
```

```python
seedlist = random.sample(range(10000), 100)
for epoch in range(epochs):
    G.train()
    count = 0
    total_loss = 0
    
    for tempseed in seedlist:
        myseed = tempseed
        myz = torch.from_numpy(np.random.RandomState(myseed).randn(1, G.z_dim)).to(device)
        myz.requires_grad = True
        
        img = G(myz, label, truncation_psi=truncation_psi)
        anchor_emb = face_encoder(img)

        positive_img = human_G(myz, label, truncation_psi=truncation_psi)
        pos_emb = face_encoder(positive_img)

        negative_img = G_toon(myz, label, truncation_psi=truncation_psi)
        neg_emb = face_encoder(negative_img)
        
        loss = triplet_loss(anchor_emb, pos_emb, neg_emb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        count += 1
    
    vis_image(G(torch.from_numpy(np.random.RandomState(28).randn(1, G.z_dim)).to(device), label, truncation_psi=truncation_psi))
    print("[Epoch: {}] loss: {}".format(epoch+1, total_loss))
```
